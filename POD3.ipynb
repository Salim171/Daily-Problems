```python
#take the height +  radius and return the volume of the cone using a function
import math

radius = float(input('Enter the Radius of cone: '))
height = float(input('Enter the Height of cone: '))

def conevolume(r, h):
  X=(1.0/3)* math.pi *math.pow(r,2) * h
  return X

z = conevolume(radius,height)
print('\n The Volume of Cone = ', z)
```
