```python

#Given a list of N fractions ,represented as two lists numerator and
#denominator the task is to determine the count of pairs of fractions whose sum equals 1

numerator = [1, 2, 2, 8]
denominator = [1, 4, 6, 12]

def calculate(n, d):
    return sum(n[i] / d[i] for i in range(len(n)))

step = 2
for i in range(0, len(numerator), step):
    x = calculate(numerator[i:i+step], denominator[i:i+step])
    print(x)
```
