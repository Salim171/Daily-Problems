```python
#Given an array of integers your task is to find the leaders in the array
import numpy as np

def findleaders(nums):
    leaders = []
    maxr = int()

    for num in reversed(nums):
        if num >= maxr:
            leaders.append(num)
            maxr = num

    leaders.reverse()
    return leaders

def main():
    array1 = np.array([16, 17, 4, 3, 5, 2])
    leaders = findleaders(array1)
    print('the leaders in this array are :',leaders)


main()
```
