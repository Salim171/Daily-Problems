```python
#given a linked list and a number k you have to reverse the part of l
#linked list with k nides and the seconde part with n-k nodes

def reverse_number_and_left(nums, num_to_find):
    if num_to_find not in nums:
        return None, "Error: The given number is not in the list."

    index = nums.index(num_to_find)

    if index > 0:
        nums[index - 1], nums[index] = nums[index], nums[index - 1]
        reversed_nums = nums[::-1]
        return reversed_nums, None
    else:
        return None, "Cannot reverse the first number in the list."

def main():
    nums = []
    for i in range(5):
        n = int(input(f"Enter number {i+1}: "))
        nums.append(n)

    num_to_find = int(input("Enter a number from the list: "))
    reversed_nums, error_message = reverse_number_and_left(nums, num_to_find)

    if error_message:
        print(error_message)
    else:
        print("Original List:", nums)
        print("Modified List:", reversed_nums)


main()
```
