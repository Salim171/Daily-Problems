```python
#user should enter an array of numbers and system return the pairs that sumup to givin number system provide it

def findpairswithsum(n, target):
    pairs = []
    set1 = set()

    for i in n:
        equation = target - i
        if equation in set1:
            pairs.append((i, equation))
        set1.add(i)

    return pairs

def main():
    list1 = []
    for i in range(5):
        n = int(input("Enter a number to sum: "))
        list1.append(n)

    n1 = 9
    pairs = findpairswithsum(list1, n1)

    if pairs:
        print("Pairs that can sum up to", n1, ":")
        for pair in pairs:
            print(pair[0], pair[1])
    else:
        print("No pairs found that sum up to", n1)

main()
```
