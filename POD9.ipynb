```python
#Given a string s the task is to count the number of substrings which contains,
#equal number of lowercase and uppercase
def find_substrings_with_equal_case(s):
    substrings = []
    n = len(s)

    for i in range(n):
        for j in range(i + 1, n + 1):
            sub_string = s[i:j]
            lower_count = sum(1 for c in sub_string if c.islower())
            upper_count = len(sub_string) - lower_count

            if lower_count == upper_count:
                substrings.append(sub_string)

    return substrings

def main():
    word = input("Enter a String (without spaces): ")
    substrings = find_substrings_with_equal_case(word)

    print("Number of substrings with equal lowercase and uppercase letters:", len(substrings))
    print("Substrings with equal lowercase and uppercase letters:", substrings)

main()
```
